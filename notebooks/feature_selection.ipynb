## Feature selection

```python
import pandas as pd
```

```python
df = pd.read_csv("data/features.csv")
df = df.iloc[:, 2:]
```

```python
y = df['label']
X = df.loc[:, df.columns != 'label']
```

Select K best

```python
from sklearn.feature_selection import SelectKBest, f_classif

print(X.shape)

selector = SelectKBest(f_classif, k=25)
X_new = selector.fit_transform(X, y)
sel_cols = X.columns.values[selector.get_support()]
print(X_new.shape)

```

```python
df_new = pd.DataFrame(X_new, columns=sel_cols)
df_new["label"] = y
df_new
```

```python
df_new.to_csv("data/features_kbest.csv")
```

Variance threshold

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.2 * (1 - .2)))
#sel = VarianceThreshold()
X_var = sel.fit_transform(X)
sel_cols = X.columns.values[sel.get_support()]
print(sel_cols)
```

```python
df_var = pd.DataFrame(X_var, columns=sel_cols)
df_var["label"] = y
df_var
```

```python
df_var.to_csv("data/features_var_thres.csv")
```
